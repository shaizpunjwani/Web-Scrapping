```python
from autoscraper import AutoScraper
```

```python
url='https://www.amazon.in/s?k=iphones'
#scraping headphones of amazon
wanted_lst=['₹39,999','Apple iPhone 11 (64GB) - (Product) RED','35,815']
scraper=AutoScraper()
#here build takes 2 parameters 1 is our url and second is the wanted_lst
result=scraper.build(url, wanted_lst)
print(result)
```

```python
scraper.get_result_similar(url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_9bc5':'Price', 'rule_puu1': 'Title', 'rule_zanx': 'Rating'})
#here we would stating only those rules which we are required a list will be passed
scraper.keep_rules(['rule_9bc5', 'rule_puu1','rule_zanx'])
scraper.save('iphone-search')
```

```python
scraper.load('iphone-search')
```
